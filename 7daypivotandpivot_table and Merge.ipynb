************************* PIVOT AND PIVOT_TABLE **********************

```python
print('example for pivot table')
import pandas as pd
data={
    'student_name':['Anu','Manu','Sumit','Anu','Mohit'],
    'class':[1,2,3,4,5],
    'rollno':[100,101,100,200,201],
    'status':['Pass', 'Fail','Pass','Pass','Fail'],
    'Percentage':['60%','35%','50%','70%','22%']
}
df=pd.DataFrame(data)
df
```

```python
df.pivot(index='class',columns='student_name',values='Percentage') #it did not display duplicate value in student_name column because index is sett by student_name
```

```python
df.pivot_table(columns='student_name',values='rollno',aggfunc='sum')
```

```python
df
df.pivot(index='class',columns='student_name',values='status')
```

```python
#create dataframe
print('example of pivot_table')
import pandas as pd
data={
    'products':['A','B','C','D','E','F','G','H','I'],
    'months':['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep'],
    'Sales':[100,300,200,400,500,345,290,390,493]
}
df=pd.DataFrame(data)
df
#use pivot_table option
table=pd.pivot_table(df,index='products',columns='months',values='Sales',aggfunc='median')
table
```

```python
import pandas as pd

data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02',
             '2023-01-03', '2023-01-03', '2023-01-04', '2023-01-04'],
    'Product': ['iPhone 13', 'Galaxy S22', 'iPad Air', 'MacBook Pro',
                'iPhone 13', 'Galaxy S22', 'iPad Air', 'MacBook Pro'],
    'Category': ['Smartphone', 'Smartphone', 'Tablet', 'Laptop',
                'Smartphone', 'Smartphone', 'Tablet', 'Laptop'],
    'Region': ['East', 'West', 'East', 'West', 'East', 'West', 'East', 'West'],
    'Sales': [1200, 1100, 800, 2500, 1200, 1100, 800, 2500],
    'Units': [10, 11, 8, 5, 10, 11, 8, 5]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime
df
```

```python
pivot_result = df.pivot(index='Date', columns='Product', values='Sales')#here date is index column and duplicate date is ommit here
pivot_result
```

```python
print('pivot table is here shows the east and west region wise sales display')
pivot_table_result = pd.pivot_table(df,
                                   values='Sales',
                                   index='Region',
                                   columns='Category',
                                   aggfunc='sum',
                                   fill_value=0)
pivot_table_result
```

******************* JOINS ARE CALLED MERGE IN PANDAS ********************
**************** FOUR TYPE OF MERGE LEFT, RIGHT, INNER AND OUTER ***********************

```python
# in pandas the joins are called merge
# types left right inner and outer
data1 = pd.DataFrame({
    'rollno':[11,2,3,4,5],
    'name':['a','b','c','d','e'],
    'marks':[10,20,30,40,50]
})
data3 = pd.DataFrame({
    'rollno':[11,22,33],
    'name':['sd','asd','asd'],
    'total':[10,20,30]
})
# inner join
inner_join = pd.merge(data1,data3,on='rollno',how='inner')
inner_join
```

```python
left_join = pd.merge(data1,data3,on='rollno',how='left')
left_join
```

```python
right_join = pd.merge(data1,data3,on='rollno',how='right')
right_join
```

```python
outer_join = pd.merge(data1,data3,on='rollno',how='outer')
outer_join
```
