*********************** TO STORE DATE IN THE DATAFRAME *************************

```python
print('date store regarding passengers')
import pandas as pd
data={
    "Name":["amit", "sumit","radha","anu"],
    "age":[22,33,32,34],
    "sex": ["male","male","female","female"]
}
df=pd.DataFrame(data)
print(df)
```

```python
print("I’m just interested in working with the data in the column Age")
df["age"]
```

```python
print("You can create a Series from scratch as well:")
ages=pd.Series([22,33,33,23], name="age")
print(ages)
```

```python
print("I want to know the maximum Age of the passengers")
print(df["age"].max())
print("find max age to the Series:")
print(ages.max())
```

```python
print("I’m interested in some basic statistics of the numerical data of my data table describe")
df.describe()
```
